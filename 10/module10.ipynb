### Importing DogsCatsClassifier class from dogs_cats file

```python
from dogs_cats import DogsCatsClassifier
```

### Creating an object to use the created DogsCatsClassifier class

```python
classifier= DogsCatsClassifier()
```

### Creating dataset folders for training, testing and Validation

```python
classifier.make_dataset_folders('train', 2400, 11999) # Creating dataset folder for training
```

```python
classifier.make_dataset_folders('validation', 0, 2399) # Creating dataset folder for Validation
```

```python
classifier.make_dataset_folders('test', 12000, 12499) # Creating dataset folder for testing
```

### Creating dataset objects for the training, testing and validation data

```python
classifier.make_dataset() # Creates the dataset objects
```

### Initializing the network

```python
classifier.build_network(augmentation=True) # Build the neural network architecture and compile it.
# The 'augmentation=True' argument enables data augmentation, which applies random transformations like flipping, rotation,
# and zooming to the input images during training. This improves model generalization and reduces the risk of overfitting.
```

### Checking the summary of the model to understand the number of trainable parameters

```python
classifier.model.summary()  # Summary of the trained model
```

### Training the model and showing the plot of both training and validation accuracy

```python
classifier.train('model.Likhitha-Kolluru.dogs-cats.keras') # Stopping training at 72% accuracy because of computational resources
```

### Predicting for a sample

```python
sample_image = 'dogs-vs-cats/test/dog/dog.12490.jpg'
```

```python
classifier.predict(sample_image)
```

### Predicting for 5 images

```python
import os
start_index = 12490
end_index = 12495
category = 'dog'

# Loop through the specified range
for i in range(start_index, end_index + 1):
    # Construct the image path
    image_file = f'dogs-vs-cats/test/{category}/{category}.{i}.jpg'

    # Check if the file exists
    if not os.path.exists(image_file):
        print(f"Image {image_file} does not exist") # Since the dataset is created, we might not know if the files exists or not
        continue

    # Predict and display results
    print(f"Predicting for {category}.{i}.jpg...")
    classifier.predict(image_file)
```
