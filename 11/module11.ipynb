
The following are implemeted in this file:
1. Sentiment analysis on the IMDb dataset using the Imdb class.
2. Text classification on the Reuters dataset using the Reuters class.
3. Regression analysis on the Boston Housing dataset using the BostonHousing class


### Importing necessary classes

```python
# Import the Imdb, Reuters, and BostonHousing classes from their respective modules.
from imdb import Imdb  # Import the Imdb class for sentiment analysis on the IMDb dataset.
from reuters import Reuters  # Import the Reuters class for text classification on the Reuters dataset.
from boston_housing import BostonHousing  # Import the BostonHousing class for regression analysis on the Boston Housing dataset.


```

### Sentiment Analysis using IMDb Dataset (Binary classification)

```python
imdb_model = Imdb()  # Create an instance of the Imdb class
```

```python
imdb_model.prepare_data()  # Load and preprocess the IMDb dataset
```

```python
imdb_model.build_model()  # Build the sentiment analysis model
```

```python
imdb_model.train(epochs=10, batch_size=512)  # Train the model
```

```python
imdb_model.plot_loss()  # Visualize the training and validation loss
```

```python
imdb_model.plot_accuracy()  # Visualize the training and validation accuracy
```

```python
imdb_model.evaluate()  # Evaluate the model on the test set
```

### Regression Analysis using Boston Housing Dataset
